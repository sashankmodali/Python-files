```python
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import clear_output
```

```python
j=40
coeff = np.zeros((j,j+1));

# print(coeff)
coeff[0,0]= 1;
coeff[1,1]= 1;
for i in range(2,coeff.shape[0]):
    coeff[i,1:(i+2)] = coeff[(i-1),:i+1]*(2*i-1)
    coeff[i,:] = (coeff[i,:] - (i-1)*coeff[i-2,:])/i;

# print(coeff)
```

```python
xstep = np.arange(-1,1,0.01)
xstep = xstep.reshape((1,xstep.shape[0])); 

values = np.zeros((coeff.shape[0],xstep.shape[1]))
for i in range(coeff.shape[1]):
    values = values + coeff[:,i,None]*xstep**i

# print(values)
```

```python
plt.figure()
i=9
for i in range(values.shape[0]):
# print(xstep[0])
# print(values[i,:])
    plt.plot(xstep[0,:],values[i,:],label=i)
    clear_output(wait=True);
    print("plotting degree : " + str(i))
plt.legend()
plt.show()
```
